```python
# Calculate whether a guy in jan month is eligible or not bonus ... def a fun 

def bonus_cal(sales_val):
    eligible_col = np.where( sales_val>10 , 'eligilbe', 'not_eligible')
    return eligible_col
df['jan_bonus'] = bonus_cal(df['Jan'])
df
```

```python
newdf = df.select_dtypes(include=['int64', 'float64']).apply(bonus_cal)
newdf
```

```python
df1 = df.copy()
res = pd.concat([df1, newdf], axis=1)
res
```
