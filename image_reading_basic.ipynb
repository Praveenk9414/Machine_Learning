```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_float
```

```python
image = io.imread("Desktop/test_image.jpg")
plt.imshow(image)
```

Summary of When to Use Each Data Type:
uint8: Most common, used for regular images with pixel values between 0 and 255.

float32: Used when more precision is needed, such as for image processing algorithms (normalization, filtering).

uint16: Used for images requiring a wider range of pixel values (medical, scientific imaging).

bool_: Used for binary masks or segmentation tasks.


```python
float_img = img_as_float(image)
dark_image = float_img*0.5
plt.imshow(dark_image)
```
