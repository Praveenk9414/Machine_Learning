```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("house_train.csv")
df.shape
```

```python
df.columns
```

```python
sns.distplot(df['LotArea'], kde=False)  # provides a histogram with distribution of lotarea in the dataset
plt.show()
```

```python
plot = sns.countplot(x="Exterior1st", data=df) # provides a barchart
# plt.show()
plot.set_xticklabels(plot.get_xticklabels(), rotation=40); # rotates the name so that it can be readable
plt.show()
```

```python
sns.regplot(x='LotArea', y='SalePrice', data=df)  # provides a scatterplot
plt.show()
```

```python
df['LotArea'].quantile([0.5, 0.95, 0.99])
```

```python
df_sub = df.loc[df['LotArea'] < df['LotArea'].quantile(0.95)]
```

```python
sns.regplot(x='LotArea', y='SalePrice', data=df_sub)
plt.show()
```

```python
df.columns
```

```python
SFcols = [x for x in df.columns if 'SF' in x]   # List Comprehesion 
# here x is rep the col_values
SFcols
```

```python
fig,axs = plt.subplots(nrows = 3,ncols = 3, figsize = (10,10))
for i in range(0, len(SFcols)):
    rows = i // 3
    cols = i % 3
    ax = axs[rows, cols]
    plot = sns.regplot(x=SFcols[i], y='SalePrice', data=df, ax=ax)
plt.show()
```

```python
fig_1,axs=plt.subplots(figsize=(10,10))
plot = sns.boxplot(x='Exterior1st', y='SalePrice', data=df, ax=axs);
plot.set_xticklabels(plot.get_xticklabels(), rotation=40);
plt.show(fig_1)
```
