```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head
```

```python
plt.xlabel('cgpa')
plt.ylabel('package')
plt.scatter(df.cgpa, df.package, color = "Green", marker = '+')
plt.plot()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['cgpa']], df[['package']])
```

```python
reg.predict([[9]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel("cpga"), plt.ylabel("package")
plt.scatter(df[['cgpa']], df[['package']], color = 'red', marker = '+')
plt.plot(df.cgpa, reg.predict(df[['cgpa']]), color = 'green')
```
