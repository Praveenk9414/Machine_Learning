```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.NaT,pd.NaT]
})
```

```python
df.head()
```

```python
# DataFrame.dropna(*, axis=0, how=_NoDefault.no_default,
#                  thresh=_NoDefault.no_default, subset=None, inplace=False, 
#                  ignore_index=False)
print(df)
df.dropna(axis=1, how="all")   # how{‘any’, ‘all’}, default ‘any’
```

```python
# DataFrame.drop_duplicates(subset=None, *, keep='first', 
#                           inplace=False, ignore_index=False)

df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
print(df,"\n")
print(df.drop_duplicates(subset=['brand'], keep='last'), "\n") #default 'first'
print(df.drop_duplicates(subset=['rating'], keep=False))
```
