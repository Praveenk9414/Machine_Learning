What is the average Price of all cars in the dataset?

```python
df1.groupby('make')['price'].mean()
```

Which is the cheapest make and costliest make of car in the lot?

```python
df2 = df1.groupby('make')['price'].agg(['max', 'min'])
df2.reset_index(inplace=True)
df2.columns = ['Company', 'Costliest_Car', 'Cheapest_Car']
df2
```

How many cars have horsepower greater than 100?

```python
(df1['horsepower']>100).value_counts()
```

How many hatchback cars are in the dataset ?

```python
(df1['body-style'] == 'hatchback').value_counts()
```

What are the 3 most commonly found cars in the dataset?

```python
df3 = df1['make'].value_counts()
df3 = df3.reset_index()
df3.columns=['make', 'val_counts']
df3.head(3)
```

Someone purchased a car for 7099, what is the make of the car?

```python
df1[df1['price'] == 7099]['make'].values[0]
```

Which cars are priced greater than 4000?

```python
df1[df1['price']>4000]['make']
```

Which are the cars that are both a sedan and priced less than 5000?

```python
df1[(df1['body-style']=='sedan') & (df1['price']<5000)]
```
