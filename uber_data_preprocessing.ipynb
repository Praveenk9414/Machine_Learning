```python
# DataFrame.drop(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

df6.drop(['Miles_cat', 'My_year',], axis=1, inplace = True)
df6
```

----------- Groupby summarize ----------

```python
# DataFrame.groupby(by=None, axis=<no_default>, level=None, as_index=True, sort=True, group_keys=True, observed=<no_default>, dropna=True)

# 1. Calculates the mean for all numeric columns for each group.
# Produces a DataFrame as output.
print(df6.groupby('START*').mean(numeric_only=True))
print(" ----------- ----------- ----------")
# 2. Calculates the mean only for the MILES* column.
# Produces a Series as output.
print(df6.groupby('START*')['MILES*'].mean())
print(" ----------- ----------- ----------")

# 3. Explicitly specifies the aggregation function.
# Equivalent to the second method but allows for customization with multiple aggregations.
print(df6.groupby('START*')['MILES*'].agg('mean'))
print(" ----------- ----------- ----------")

# DataFrame.agg(func=None, axis=0, *args, **kwargs)
print(df6.groupby('START*')['MILES*'].agg(['sum', 'mean']))
```

--------------- In class assignment -------------

Find most recent & the earliest travelled date and mean distance traveled for each start city

Now here recent --> max(START_DATE*) and earliest --> min(START_DATE*)

```python
df7 = df.copy()
df7.tail()
```

```python
df7.drop(1155, axis=0, inplace=True)
df7.tail()
```

```python
df7['START_DATE*'] = pd.to_datetime(df7['START_DATE*'])
df7['START_DATE*'].info()
```

```python
df7.groupby('START*')['START_DATE*'].agg(['max', 'min'])  # NOW this much can be done if u want to apply agg on a single col

# But i have to apply .agg function on two col so i will have to use dictionary -> change in syntax

df8 = df7.groupby('START*').agg({'START_DATE*':['max', 'min'], 'MILES*':['mean']}) # here we first call .agg function and then assign using {key-value} pair
```

```python
df8.reset_index(inplace=True)
df8.columns = ['START_LOC', 'LATEST_TRIP', 'EARLIEST_TRIP', 'AVG_DIST']
```

------------- List Comprehension array  ------------

```python
arr = []
temp = [1,2,3,4]
for i in temp:
    arr.append(i*i)
print(arr)
```

```python
# now this can we done in one line using list comprehension

# ['opr' for 'var' in 'arr']
arr_1 = [x*x for x in temp]
print(arr_1)
```
